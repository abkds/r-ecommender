```python
from collections import defaultdict
user_item_matrix = defaultdict(lambda: defaultdict(int))
with open("final_interactions_count") as f:
    for line in f.readlines():
        line = line.split()
        user_item_matrix[line[0]][line[1]] = 1
        

    
```

```python
sum_total=0
for e in user_item_matrix:
    for item in user_item_matrix[e]:
        sum_total += user_item_matrix[e][item]
```
